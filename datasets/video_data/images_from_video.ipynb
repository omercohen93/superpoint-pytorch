# Change vid to grayscale and sve images

```
import cv2 
  
# reading the vedio 
cap = cv2.VideoCapture(path+'/datasets/video_data/vid3.mp4') 
n_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'MJPG')
out = cv2.VideoWriter(path+'/datasets/video_data/gray_vid1.mp4', fourcc, fps, (w, h), isColor=False)
im_path = path+'/datasets/video_data/video3_images/'


 
# running the loop 
ret, img = cap.read()
print(ret)
count=0
while ret: 
      
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
    out.write(gray)
    cv2.imwrite(im_path+"frame{}.jpg".format(count), gray)     # save frame as JPEG file      
    ret,img = cap.read()
    print('Read a new frame (fram {}): '.format(count), ret)  
    count+=1
# closing the window 
cv2.destroyAllWindows() 
cap.release()
```
